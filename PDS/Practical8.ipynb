 A) Draw graph for the given list of nodes and edges using NetworkX.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph(directed=True)
G.add_edges_from(
    [('A', 'B'), ('A', 'C'), ('D', 'B'), ('E', 'C'), ('E', 'F'),
     ('B', 'H'), ('B', 'G'), ('B', 'F'), ('C', 'G')])

val_map = {'A': 1.0,
           'D': 0.5,
           'H': 0.0}

values = [val_map.get(node, 0.25) for node in G.nodes()]
options = {
    'node_color': 'yellow',
    'node_size': 500,
    'width': 3,
    'arrowstyle': '-|>',
    'arrowsize': 15,
}
nx.draw_networkx(G, arrows=True, **options)
plt.show()
```
