```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "./data/loan_data_set.csv"
data = pd.read_csv(file_path)
df  = pd.DataFrame(data)
```

A) Applicant income is above 5000 and property area is Urban

```python
gr_a = df.groupby('ApplicantIncome').apply(lambda df: df[(df['ApplicantIncome'] > 5000) & (df['Property_Area'] == 'Urban')])
# gr_a.head()
```

B) Coapplicant income is above 10000 and property area is Urban

```python
gr_b = df.groupby('CoapplicantIncome').apply(lambda df: df[(df['CoapplicantIncome'] > 10000) & (df['Property_Area'] == 'Urban')])
# gr_b.head()
```

C) Coapplicant income is above 5000 and property area is Rural


```python
gr_c = df.groupby('CoapplicantIncome').apply(lambda df: df[(df['CoapplicantIncome'] > 5000) & (df['Property_Area'] == 'Rural')])
# gr_c.head()
```

D) Applicant income is below 5000 and property area is Semiurban

```python
gr_d = df.groupby('ApplicantIncome').apply(lambda df: df[(df['ApplicantIncome'] < 5000) & (df['Property_Area'] == 'Semiurban')])
# gr_d.head()
```

```python
#plot the graph for visual statistics
lables = ["(A)", "(B)", "(C)", "(D)"]
values = [len(gr_a), len(gr_b), len(gr_c), len(gr_d)]
bar_color = ["#00203FFF", "#ADEFD1FF", "#2D2926FF", "#E94B3CFF"]
for i in range(len(lables)):
    plt.bar(lables[i], values[i], color=bar_color[i])
plt.xlabel("Groups")
plt.ylabel("Number of Applicants")
plt.title("Groups based on Income and Property Area")
plt.legend(loc="upper left", title="Groups", labels=[f"(A)-{len(gr_a)}", f"(B)-{len(gr_b)}", f"(C)-{len(gr_c)}", f"(D)-{len(gr_d)}"],)
plt.show()
```
