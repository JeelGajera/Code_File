```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_theme(style="whitegrid")

file_path = "./data/loan_data_set.csv"
data = pd.read_csv(file_path)
df  = pd.DataFrame(data)
```

A) Bar Plot â€“ Plot the number of male/female applicants.

```python
data = df['Gender']
sns.barplot(df, x=data.value_counts().index, y=data.value_counts().values, hue=data.value_counts().values).set_title('Number of Applicants Vs Gender')
```

B) Strip Plot â€“ Plot this with respect to the Applicant Income by grouping according to their Property area.

```python
sns.stripplot(df, x=df['ApplicantIncome'], y=df['Property_Area'], hue=df['Property_Area']).set_title('Applicant Income Vs Property Area')
```
