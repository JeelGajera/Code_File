```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

file_path = "./data/US_Superstore_data.xls"
data = pd.read_excel(file_path)
df  = pd.DataFrame(data)
```

```python
store_data = df[['Order Date', 'Category', 'Sales']]
sns.lineplot(data=store_data, x="Order Date", y="Sales", hue="Category")
```
