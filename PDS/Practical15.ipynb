```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "./data/loan_data_set.csv"
data = pd.read_csv(file_path)
df  = pd.DataFrame(data)
numeric_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']
```

A) Display the distribution of the dataset on dataset wise and attribute wise. Use scatter plot, histogram and apply t-test on it.

```python
#plot hiatogram
plt.figure(figsize=(20,10))
i=1
for col in numeric_cols:
    plt.subplot(2,3,i)
    sns.histplot(df[col], kde=True, color='blue', bins=10).set_title(col)
    sns.histplot(df[col],cumulative=True, kde=True, color='red', bins=10)
    i+=1
```

```python
#plot scatter plot
sns.scatterplot(data=df, x="ApplicantIncome", y="LoanAmount", hue="Gender", style="Loan_Status")
```

```python
sns.scatterplot(data=df, x="ApplicantIncome", y="LoanAmount", hue='Education', style="Loan_Status")
```

```python
sns.scatterplot(data=df, x="ApplicantIncome", y="LoanAmount", hue='Self_Employed', style="Gender")
```

```python
# perform t-test
df.dropna(inplace=True) #clean dataset
print(ttest_ind(df['ApplicantIncome'], df['CoapplicantIncome']))
print(ttest_ind(df['ApplicantIncome'], df['LoanAmount']))
print(ttest_ind(df['ApplicantIncome'], df['Loan_Amount_Term']))
print(ttest_ind(df['ApplicantIncome'], df['Credit_History']))
```
